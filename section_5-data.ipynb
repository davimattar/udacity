```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Excel file into a pandas DataFrame
df = pd.read_excel('section-5-data.xlsx')
```

```python
df
```

```python
# Convert 'Unnamed: 0' to datetime and set as index
df['Unnamed: 0'] = pd.to_datetime(df['Unnamed: 0'])
df.set_index('Unnamed: 0', inplace=True)

# Plotting the data
plt.figure(figsize=(15, 8))

# Plot each event type as a separate line
plt.plot(df.index, df['Choose Car'], label='Choose Car')
plt.plot(df.index, df['Search'], label='Search')
plt.plot(df.index, df['Open'], label='Open')
plt.plot(df.index, df['Begin Ride'], label='Begin Ride')
plt.plot(df.index, df['Request Car'], label='Request Car')

# Adding title and labels
plt.title('Number of Events of Each Type Per Day')
plt.xlabel('Date')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
